```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL="https://www.freelancer.in/jobs/data-analytics/"
r=requests.get(URL)
```

```python
print(r)
```

```python
soup=BeautifulSoup(r.content,'html5lib')
print(soup.prettify())
```

```python
quote={"Job":[],"Description":[],"Requirements":[],"Bid Price":[]}
```

Job Name:
JobSearchCard-primary-heading-link
Description:
JobSearchCard-primary-description
Requirements:
JobSearchCard-primary-tagsLink
Bid Price:
JobSearchCard-primary-price

```python
table=soup.find('div',attrs={'id':"project-list"})
```

```python
table
```

```python
for row in table.findAll("div",attrs={'class':'JobSearchCard-primary-heading'}):
    print(row)
    quote['Job'].append(row.a['href'])
```

```python
for row in table.findAll("p",attrs={'class':'JobSearchCard-primary-description'}):
    quote['Description'].append(row.get_text().strip())
```

```python
for row in table.findAll("div",attrs={'class':'JobSearchCard-primary-tags'}):
    quote['Requirements'].append(row.a['href'])
```

```python
for row in table.findAll("div",attrs={'class':'JobSearchCard-primary-price'}):
    quote['Bid Price'].append(row.text.strip())
```

```python
quote
df=pd.DataFrame(quote)
df.to_csv("df.csv")
```
